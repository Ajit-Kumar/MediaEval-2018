# **Caption Pre-Processing**



```
import string
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer

tokenizer = RegexpTokenizer(r'\w+')
porter = PorterStemmer()

def text_process(caption):

  """
  1. Removing punctuation mark
  2. Removing stopwords
  3. Normalizing Case
  4. Stemming
  5. return clean text
  """
    # Check characters to see if they are in punctuation and normalizing case
  nopunc = [char.lower for char in caption if char not in string.punctuation] 
  nopunc = tokenizer.tokenize(caption)

  # remove any stopwords 
  cleaned_caption = [porter.stem(word) for word in nopunc if word.lower() not in stopwords.words('english')]

  # Join the characters again to form the string.
  cleaned_caption = ' '.join(cleaned_caption)

  return cleaned_caption

#applying the text_process function on caption column of dev set 
caption['cleaned caption'] = caption['caption'].apply(text_process)
#applying the text_process function on caption column of dev set 
caption_test['cleaned caption'] = caption_test['caption'].apply(text_process)
```

## **Spearman Score Calculation Method**

```
def Get_score(Y_pred,Y_true):
    '''Calculate the Spearmann"s correlation coefficient'''
    Y_pred = np.squeeze(Y_pred)
    Y_true = np.squeeze(Y_true)
    if Y_pred.shape != Y_true.shape:
        print('Input shapes don\'t match!')
    else:
        if len(Y_pred.shape) == 1:
            Res = pd.DataFrame({'Y_true':Y_true,'Y_pred':Y_pred})
            score_mat = Res[['Y_true','Y_pred']].corr(method='spearman',min_periods=1)
            print('The Spearman\'s correlation coefficient is: %.3f' % score_mat.iloc[1][0])
        else:
            for ii in range(Y_pred.shape[1]):
                Get_score(Y_pred[:,ii],Y_true[:,ii])
```

## **Implementing LSTM Model**

```
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense, LSTM, MaxPooling1D, Dropout, Activation
from keras.layers.embeddings import Embedding

#function for LSTM
def lstm(source,target):

  maxlen = 20
  vocabulary_size = 25000
  epochs  = 3

  tokenizer = Tokenizer(num_words= vocabulary_size)
  tokenizer.fit_on_texts(source)
  sequences = tokenizer.texts_to_sequences(source)
  data = pad_sequences(sequences, maxlen=maxlen)
  vocab_size = len(tokenizer.word_index) + 1

  #splitting the data in test and train
  X_train_lstm, X_test_lstm, y_train_lstm, y_test_lstm = train_test_split(data, target, test_size=0.2, random_state=101)

  model_lstm = Sequential()
  model_lstm.add(Embedding(vocab_size, 100, input_length=maxlen))
  model_lstm.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
  model_lstm.add(Dense(2, activation='sigmoid'))
  model_lstm.compile(loss='binary_crossentropy', 
                     optimizer='adam', 
                     metrics=['accuracy'])

  # fitting the model
  model_lstm.fit(X_train_lstm,y_train_lstm, epochs=epochs)

  #testing the model
  lstm_output = model_lstm.predict(X_test_lstm)
  lstm_score  = Get_score(lstm_output,y_test_lstm)

  return lstm_output
```

## **Implementing SVR**

```
from sklearn.svm import SVR

def svr_c3d(source,target):

  X_train_svr, X_test_svr, y_train_svr, y_test_svr = train_test_split(source,target, test_size=0.2,random_state=46)

  svr_c3d = SVR (kernel='linear')

  svr_c3d.fit(X_train_svr, y_train_svr)
  svr_output = svr_c3d.predict(X_test_svr)
  Get_score(svr_output,y_test_svr)
  
  return svr_output
```

## **Implementing Random Forest**

```
from sklearn.ensemble import RandomForestRegressor 
from sklearn.model_selection import KFold

def RF_c3d(source,target,test):

  RF_c3d = RandomForestRegressor(n_estimators=100,max_features='sqrt',random_state=10)
  X = source
  y = target
  
  # KFold Cross Validation approach
  kf = KFold(n_splits=5,shuffle=False)
  kf.split(X) 

  for train_index, test_index in kf.split(X):

    # Split train-test
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # Train the model
    model = RF_c3d.fit( X_train,y_train)
  
  rf_output_c3d = RF_c3d.predict(X_test)
  
  rf_output_c3d_test = RF_c3d.predict(test)

  Get_score(rf_output_c3d,y_test)
  return rf_output_c3d,rf_output_c3d_test
```

```
#for short tem 
print('C3D short term memorability using Random Forest')
rf_c3d_sc_output,rf_c3d_sc_output_test = RF_c3d(ground_truth_c3d.iloc[:, 5:],ground_truth_c3d['short-term_memorability'],C3D_test_df.iloc[:,:-1])
#for long tem 
print('C3D long term memorability using Random Forest')
rf_c3d_lc_output = RF_c3d(ground_truth_c3d.iloc[:, 5:],ground_truth_c3d['long-term_memorability'],C3D_test_df.iloc[:,:-1])
```

```
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

def asthetic(model,source,target,test):
  """
  1 . RF for RandomForestRegressor
  2 . bay for BayesianRidge
  3 . svr for SVR
  4 . dtr for DecisionTreeRegressor   
  """
  #ipdb.set_trace()
  if model == 'RF':

    rf_asthetic = RandomForestRegressor(n_estimators = 400,max_features = 'sqrt',random_state=105)
    X_train_rf_af, X_test_rf_af, y_train_rf_af, y_test_rf_af = train_test_split(source,target, test_size=0.2,random_state=101)

    rf_asthetic.fit(X_train_rf_af,y_train_rf_af)
    output_aesthetic = rf_asthetic.predict(X_test_rf_af)
    output_aesthetic_test = rf_asthetic.predict(test)
    Get_score(output_aesthetic,y_test_rf_af)

  elif model == 'bay':

    bay_ridge1 = BayesianRidge()

    X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(source,target, test_size=0.2,random_state=80)

    bay_ridge1.fit(X_train_b,y_train_b)
    output_aesthetic = output_aesthetic_test = bay_ridge1.predict(X_test_b)
    Get_score(output_aesthetic,y_test_b)

  elif model == 'svr':

    svr_af = SVR(kernel='linear')

    X_train_saf, X_test_saf, y_train_saf, y_test_saf = train_test_split(source,target, test_size=0.2,random_state=101)

    svr_af.fit(X_train_saf,y_train_saf)
    output_aesthetic = output_aesthetic_test = svr_af.predict(X_test_saf)
    Get_score(output_aesthetic,y_test_saf)

  elif model == 'dtr':

    dtr = DecisionTreeRegressor()

    X_train_dtr, X_test_dtr, y_train_dtr, y_test_dtr = train_test_split(source,target, test_size=0.2,random_state=42)

    dtr.fit(X_train_dtr,y_train_dtr)
    output_aesthetic = output_aesthetic_test = dtr.predict(X_test_dtr)
    Get_score(output_aesthetic,y_test_dtr)


  elif model == 'lr':

    lr_af = LinearRegression()

    X_train_lraf, X_test_lraf, y_train_lraf, y_test_lraf = train_test_split(source,target, test_size=0.2,random_state=80)

    lr_af.fit(X_train_lraf,y_train_lraf)
    output_aesthetic = lr_af.predict(X_test_lraf)
    output_aesthetic_test = lr_af.predict(test)
    Get_score(output_aesthetic,y_test_lraf)

  return output_aesthetic,output_aesthetic_test
```
